**Découverte jeu de données Benchmark 1**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_with_anom(data) :
    anom = data[data["is_anomaly"] == 1]
    plt.plot(data["timestamp"], data["value"])
    plt.scatter(anom["timestamp"], anom["value"], color="red", marker='o')
    plt.show()
```

```python
for i in range(60,66) :
    data = pd.read_csv("real_" + str(i) + ".csv")
    plot_with_anom(data)
```

Longueur des séries :

```python
long = [pd.read_csv("real_" + str(i) + ".csv").shape[0] for i in range(1,68)]
plt.hist(long)
plt.show()
```

Chaque série contient un peu plus de 1400 valeurs, sauf quelques unes qui n'en ont que 700.

Nombre de séries sans aucune anomalie :

```python
cpt = 0
for i in range(1,68) :
    data = pd.read_csv("real_" + str(i) + ".csv")
    cpt += int(data[data["is_anomaly"] == 1].empty)
    
print(cpt)
```
