**Découverte jeux de données**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Données benchmark 1**

```python
path = "A1Benchmark/"
```

```python
serie_1 = pd.read_csv(path + "real_1.csv", index_col=0)
serie_1.head()
```

```python
def plot_with_anom(data) :
    anom = data[data["is_anomaly"] == 1]
    plt.plot(data["timestamp"], data["value"])
    plt.scatter(anom["timestamp"], anom["value"], color="red", marker='o')
    plt.show()
```

```python
for i in range(60,66) :
    data = pd.read_csv(path + "real_" + str(i) + ".csv")
    plot_with_anom(data)
```

Longueur des séries :

```python
long = [pd.read_csv(path + "real_" + str(i) + ".csv").shape[0] for i in range(1,68)]
plt.hist(long)
plt.show()
np.unique(long)
```

Presque toutes les séries contiennent entre 1420 et 1461 valeurs, sauf quelques unes qui n'en ont que 741.

Nombre de séries sans aucune anomalie :

```python
cpt = 0
for i in range(1,68) :
    data = pd.read_csv(path + "real_" + str(i) + ".csv")
    cpt += int(data[data["is_anomaly"] == 1].empty)
    
print(cpt)
```
